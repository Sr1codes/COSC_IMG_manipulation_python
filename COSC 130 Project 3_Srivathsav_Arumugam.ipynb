We will start by first understanding how colored images exist

Depending on the information represented by each pixel, images can be divided into binary images, grayscale images, RGB images, etc.


RGB Image
In RGB or color images, the information for each pixel requires an array of numbers to represent. So we need a three-dimensional matrix to represent an image. Almost all colors in nature can be composed of three colors: red (R), green (G), and blue (B). So each pixel can be represented by a red/green/blue array in an RGB image.
RGB color model stores individual values for red, green, and blue. With a color space based on the RGB color model, the three primaries are added together to create colors from completely white to completely black.

Grayscale Image
The grayscale image adds a color depth between black and white in the binary image to form a grayscale image. Such images are usually displayed as grayscales from the darkest black to the brightest white, and each color depth is called a grayscale.

Binary Image:
In a binary image, the pixel value is represented by a 0 or 1. Generally, 0 is for black and 1 is for white. 

Each separate color plane whether red, green or blue, when looked at individually looks like a grayscale image.

# RGB to Grayscale converson

There are a number of commonly used methods to convert an RGB image to a grayscale image, such as average method and weighted method.

Average Method
The Average method takes the average value of R, G, and B as the grayscale value.

Grayscale = (R + G + B ) / 3
A minor note: Theoretically, the formula is 100% correct. But when writing code, you may encounter uint8 overflow error — the sum of R, G, and B is greater than 255. To avoid the exception, R, G, and B should be calculated, respectively.

Grayscale = R / 3 + G / 3 + B / 3
The average method is simple but doesn’t work as well as expected. The reason is that human eyeballs react differently to RGB. Eyes are most sensitive to green light, less sensitive to red light, and the least sensitive to blue light. Therefore, the three colors should have different weights in the distribution. That brings us to the weighted method.

The Weighted Method
The weighted method, also called the luminosity method, weighs red, green, and blue according to their wavelengths. The improved formula is as follows:

Grayscale = 0.299R + 0.587G + 0.114B

## Task 3: Function 3: rgb_to_grayscale_WAVG() using weighed averaging : 10 points

Keeping everything else the same, this time use the weighed average formula to calculate the grayscale value:
 0.299R + 0.587G + 0.114B
 
Your function should be named rgb_to_grayscale_WAVG. The code to call the function is given to you. 

```python
img = Image.open('cat.jpg')
img_arr = np.asarray(img)

'''CREATE YOUR FUNCTION HERE'''
def rgb_to_grayscale_WAVG(rgb_img_arr):
    width, height, _ = rgb_img_arr.shape
    grayscale_img = np.zeros((width, height))
    for x in range(width):
        for y in range(height):
            r, g, b = rgb_img_arr[x, y]
            grayscale_value = int(0.299 * r + 0.587 * g + 0.114 * b)
            grayscale_img[x, y] = grayscale_value

    return Image.fromarray(grayscale_img)

gray_wavg = rgb_to_grayscale_WAVG(img_arr)
gray_wavg = np.asarray(gray_wavg)
gray_wavg = gray_wavg.astype(int)
gray_wavg = Image.fromarray(gray_wavg)
plt.imshow(gray_wavg)
```

```python
img = gray_wavg
img_arr = np.asarray(img)

'''CREATE YOUR FUNCTION HERE'''

def gray_to_binary(gray_img):
    width, height = gray_img.shape  
    binary_img = np.zeros((width, height))  
    
    for x in range(width):
        for y in range(height):
            pixel_intensity = gray_img[x, y]  
            binary_img[x, y] = 255 if pixel_intensity > 128 else 0  
    
    return binary_img

binary = gray_to_binary(img_arr)  
binary = np.asarray(binary)
binary = binary.astype(int)
binary = Image.fromarray(binary)
plt.imshow(binary)
```

```python
img = Image.open('cat.jpg')
img_arr = np.asarray(img)


def rgb_brightness(rgb_img, factor):
    if factor > 1 or factor < 0:
        return

    width, height, planes = rgb_img[:, :, :].shape
    augmented_img = np.zeros((width, height, planes))

    for x in range(width):
        for y in range(height):
            r = rgb_img[x, y, 0]
            g = rgb_img[x, y, 1]
            b = rgb_img[x, y, 2]
            
            aug_r = int(r * factor)
            aug_g = int(g * factor)
            aug_b = int(b * factor)
            
            augmented_img[x, y, 0] = aug_r
            augmented_img[x, y, 1] = aug_g
            augmented_img[x, y, 2] = aug_b

    return augmented_img

try:
    augmented = rgb_brightness(img_arr, 0.5)
    augmented = np.asarray(augmented)
    augmented = augmented.astype(int)
    Image.fromarray((augmented).astype(np.uint8))
    plt.imshow(augmented)

except:
    print('The factor value you entered is invalid. Please enter a value between 0 and 1')
```
